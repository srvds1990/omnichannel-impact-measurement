```python
import pandas as pd
import numpy as np

np.random.seed(42)
```

```python
# Number of HCP's to generate 
n = 10000

# HCp ID genrations 
hcp_id = range(100001, 110001)
# Distributions 
specialties = np.random.choice(
    ["Oncology", "Cardiology", "Neurology", "Endocrinology"],
    size=n,
    p=[0.45, 0.20, 0.20, 0.15]
)

#HCP Tier 
tier = np.random.choice(
    ["A", "B", "C"],
    size=n,
    p=[0.2, 0.5, 0.3]
)
```

```python
#Historical Engagements 


historical_email_opens = np.random.poisson(8, n)

historical_msl_calls = np.random.poisson(2, n)

digital_engagement_score = np.random.normal(
    60,
    15,
    n
)

digital_engagement_score = np.clip(
    digital_engagement_score,
    0,
    100
)

# YOP 

years_practice = np.random.randint(
    2,
    35,
    n
)

#Previous Scientific Activity 

publication_downloads = np.random.poisson(3, n)

webinar_attendance = np.random.poisson(2, n)



```

```python
#Treatment Assignment Logic
#This is very important.

#Real campaigns are not random.

#More engaged HCPs receive more campaigns.



probability = (
    0.01 * digital_engagement_score +
    0.05 * historical_msl_calls +
    0.03 * historical_email_opens
)

probability = 1 / (1 + np.exp(-(probability - 2)))

treatment = np.random.binomial(
    1,
    probability
)


```

```python
#Scientific Engagement Score

scientific_score = (
    digital_engagement_score * 0.4 +
    historical_email_opens * 2 +
    historical_msl_calls * 4 +
    publication_downloads * 3 +
    webinar_attendance * 5
)
# Add Campaign Impact 
scientific_score += treatment * 8
```

```python
# ADD Noise 

scientific_score += np.random.normal(
    0,
    5,
    n
)
```

```python
hcp_df = pd.DataFrame({

    "hcp_id": hcp_id,
    "specialty": specialties,
    "tier": tier,
    "years_practice": years_practice,

    "historical_email_opens": historical_email_opens,
    "historical_msl_calls": historical_msl_calls,

    "digital_engagement_score":
        digital_engagement_score,

    "publication_downloads":
        publication_downloads,

    "webinar_attendance":
        webinar_attendance,

    "treatment":
        treatment,

    "scientific_score":
        scientific_score
})
```
